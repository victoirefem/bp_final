```python
from tensorflow.keras.layers import Input, BatchNormalization, Dense, Flatten
from tensorflow.keras import Model
import numpy as np
```

```python
inputs = Input(shape=(5,5,3))
out = BatchNormalization()(inputs)
out = Flatten()(out)
out = Dense(1)(out)
model = Model(inputs, out)
```

```python
model.summary()
```

```python
X_train = np.random.rand(100,5,5,3)
X_train
```

```python
y_train = np.random.rand(100,1)
y_train
```

```python
model.compile("adam", "mse")
model.fit(X_train, y_train, epochs=10)
```

```python
model.layers[1].weights
```

```python
X = np.random.rand(1,5,5,3)
X
```

```python
new_model = Model(inputs, model.layers[1].output)
y = new_model.predict(X)
y
```

```python
gamma = model.layers[1].weights[0].numpy()
beta = model.layers[1].weights[1].numpy()
moving_mean = model.layers[1].weights[2].numpy()
moving_var = model.layers[1].weights[3].numpy()
epsilon = model.layers[1].epsilon
```

```python
a = gamma/(moving_var+epsilon)**.5
b = beta-gamma*moving_mean/(moving_var+epsilon)**.5
```

```python
y-(gamma*(X-moving_mean)/((moving_var+epsilon)**.5)+beta)
```

```python
y-(a*X+b)
```

```python
X_in = [[[int(X[0][i][j][k]*2**10) for k in range(3)] for j in range(5)] for i in range(5)]
X_in_unchanged = [[[float(X[0][i][j][k]) for k in range(3)] for j in range(5)] for i in range(5)]
a_in = [int(a[i]*2**10) for i in range(3)]
a_in_unchanged = [float(a[i]) for i in range(3)]
b_in = [int(b[i]*2**20) for i in range(3)]
b_in_unchanged = [float(b[i]) for i in range(3)]
```

```python
def BatchNormalizationInt(nRows, nCols, nChannels, n, X_in, a_in, b_in):
    X = [[[str(X_in[i][j][k]) for k in range(nChannels)] for j in range(nCols)] for i in range(nRows)]
    A = [str(a_in[k]) for k in range(nChannels)]
    B = [str(b_in[k]) for k in range(nChannels)]
    out = [[[None for _ in range(nChannels)] for _ in range(nCols)] for _ in range(nRows)]
    for i in range(nRows):
        for j in range(nCols):
            for k in range(nChannels):
                out[i][j][k] = (X_in[i][j][k]*a_in[k] + b_in[k])
                out[i][j][k] = str(out[i][j][k] // n)
    return X, A, B, out
```

```python
X_in, a_in, b_in, out = BatchNormalizationInt(5, 5, 3, 2**10, X_in, a_in, b_in)
out
```
